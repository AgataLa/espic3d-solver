```julia
using FFTW
using LinearAlgebra

# Parametry siatki
N = 128  # Liczba punktów siatki w każdym wymiarze
L = 1.0  # Rozmiar domeny
h = L / N  # Krok siatki
epsilon_0 = 8.854e-12  # Przenikalność elektryczna próżni

# Siatka współrzędnych
x = range(0, stop=L, length=N)
y = range(0, stop=L, length=N)
z = range(0, stop=L, length=N)
X, Y, Z = [repeat(reshape(range, N, 1, 1), 1, N, N) for range in (x, y, z)]

# Gęstość ładunku ρ(x, y, z)
ρ = sin.(2π .* X) .* sin.(2π .* Y) .* sin.(2π .* Z)

# FFT gęstości ładunku
ρ_hat = rfft(ρ)

# Częstości przestrzenne
kx = fftfreq(N, 1/N) .* 2π
ky = fftfreq(N, 1/N) .* 2π
kz = fftfreq(N, 1/N) .* 2π
KX, KY, KZ = [repeat(reshape(freq, N, 1, 1), 1, N, N) for freq in (kx, ky, kz)]

# Rozwiązywanie równania Poissona w przestrzeni Fouriera
ϕ_hat = similar(ρ_hat)
for i in 1:size(ϕ_hat, 1), j in 1:N, k in 1:N
    k_sq = KX[i, 1, 1]^2 + KY[1, j, 1]^2 + KZ[1, 1, k]^2
    if k_sq != 0
        ϕ_hat[i, j, k] = ρ_hat[i, j, k] / (epsilon_0 * k_sq)
    else
        ϕ_hat[i, j, k] = 0
    end
end

# iFFT potencjału
ϕ = irfft(ϕ_hat, N)

# Teraz ϕ zawiera rozwiązanie równania Poissona
ϕ_real = real.(ϕ)  # Przejście na część rzeczywistą
```
