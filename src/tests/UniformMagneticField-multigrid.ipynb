```julia
include("../pic3d.jl")
using .PIC3D
using PyPlot


PIC3D.B_ex[3] = 6.82272e-5
electron = PIC3D.Species(q=PIC3D.q_e, m=PIC3D.m_e)
push!(electron.x, [0.75, 0.5, 0.0])
push!(electron.v, [0.0, 3.0e6, 0.0])

timesteps = 1000
historyxx = []
push!(historyxx, [0.75, 0.5, 0.0])

for i = 1:timesteps
    println(i)
    PIC3D.timestep!(electron, 1)
    push!(historyxx, [electron.x[1]...])
end
```

```julia
v = 3e6
B = 6.82272e-5
q_e = -1.602176634e-19
m_e = 9.1093837139e-31 
r = abs((m_e*v) / (q_e*B))
w = -(q_e*B)/m_e
xx_an = []
push!(xx_an, [0.75, 0.5])
dt = 1e-9 
for t in 1:100
    t *= dt
    k = w*t
    x = r * cos(k) - r
    y = r * sin(k)
    push!(xx_an, [0.75, 0.5] + [x, y])
end
```

```julia
plt.figure(figsize=(8,8))
plt.plot([x[1] for x in historyxx], [x[2] for x in historyxx],linestyle="solid")
plt.plot([x[1] for x in xx_an], [x[2] for x in xx_an],linestyle="dashed")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.axis("square")
plt.grid()
plt.xticks(0.2:0.05:0.8)
plt.yticks(0.2:0.05:0.8)
;
```
